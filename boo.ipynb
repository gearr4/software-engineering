```python
#first requirement
import string
import os

file_path = 'boo.txt'

  
text = open("boo.txt", "r")

# check if size of file is 0
if os.stat(file_path).st_size == 0:
    text_file = open("boo.txt", "w")
    text = text_file.write(input("ENTER THE PARAGRAPH:"))
    text_file.close()
    print(text)

d = dict()
print(text)
text = open("boo.txt", "r")
```

```python
for line in text:
    line = line.strip()
    line = line.lower()
    line = line.translate(line.maketrans("", "", string.punctuation))
    words = line.split(" ")
    for word in words:
        if word in d:
            d[word] = d[word] + 1
        else:
            d[word] = 1
  

for key in list(d.keys()):
    print(key, ":", d[key])
```

```python
#2nd requirement
text = open("boo.txt", "r")
sword=input("enter word that need to be searched:")
rword=input("enter word to replace sword:")
for line in text:
    line = line.strip()
    line = line.lower()
    line = line.translate(line.maketrans("", "", string.punctuation))
    words = line.split(" ")
    l2 =[]
    for item in words:
        if item==sword:
            l2.append(item.replace(sword,rword))
        else:l2.append(item)
text=" ".join(l2)
f = open("boo.txt", "w")
f.write(text)
f.close()
```
