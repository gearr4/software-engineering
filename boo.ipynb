```python
#2nd requirement
text = open("boo.txt", "r")
sword=input("enter word that need to be searched:")
rword=input("enter word to replace sword:")
for line in text:
    line = line.strip()
    line = line.lower()
    line = line.translate(line.maketrans("", "", string.punctuation))
    words = line.split(" ")
    print(words)
    l2 =[]
    for item in words:
        if item==sword:
            l2.append(item.replace(sword,rword))
            print("word replacement done")
        else:l2.append(item)
text=" ".join(l2)
f = open("boo.txt", "w")
f.write(text)
f.close()
```
